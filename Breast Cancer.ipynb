#### Spliting the data set Train and test 

```python
data.head()
```

```python
##Independent Variable
X= data.iloc[:,1:-1].values
```

```python
X.shape
```

```python
#target variable
y = data.iloc[:,-1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

#### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### Building The Model

##### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score,recall_score
```

```python
acc = accuracy_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
```

```python
result = pd.DataFrame([['LogisticRegression',acc,f1,prec,rec]],
                     columns=['Model','accuracy','f1_score','precision','recall'])
```

```python
result
```

```python
Confu_m=confusion_matrix(y_test,y_pred)
```

```python
print(Confu_m)
```

### Cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuraciess = cross_val_score(estimator=clf,X=X_train,y=y_train,cv=10 )
```

```python
print('Accuracy is {:.2f}%'.format(accuraciess.mean()*100))
print('Standard Deviation is {:.2f}%'.format(accuraciess.std()*100))
```

### Final Model

```python
clf = LogisticRegression(C=0.5, random_state=0, solver='liblinear')
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
acc = accuracy_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)

mod_result = pd.DataFrame([['Final Logistic Regression',acc,f1,prec,rec]],
                     columns=['Model','accuracy','f1_score','precision','recall'])
```

```python
results = result.append(mod_result,ignore_index=True)

results
```

### Cross validation

```python
accuraciess = cross_val_score(estimator=clf,X=X_train,y=y_train,cv=10 )

print('Accuracy is {:.2f}%'.format(accuraciess.mean()*100))
print('Standard Deviation is {:.2f}%'.format(accuraciess.std()*100))
```
